1. Importing Libraries

```python
import pandas as pd #Data Manipulation
import matplotlib.pyplot as plt #Data Visualization
```

2. Loading Dataset

---



```python
data = pd.read_csv('hardwareStore.csv')
data.head()
```

3. Data Loading & Initial Analysis
---

```python
statistics = data.describe
```

4. Data Manipulation
---

```python
product_revenue = data.groupby('STANDARD_COST').sum()
highest_revenue_product = product_revenue.idxmax()
average_quantity_per_day = highest_revenue_product.mean()

print(product_revenue)
```

5. Data Visualization
---

```python
product_revenue.plot(kind = 'bar')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.title('Revenue by Product')
plt.xticks(rotation = 45)
plt.show
```

6. Analysis Report
---

```python
report = """
Analysis Report

1. Data Loading and Initial Analysis:
    Basic statistics of the dataset:
    {}
2. Data Manipulation:
    Total revenue for each product:
    {}
    Product with the highest revenue: {}
    Average quantity sold per day{:.2f}
3. Data Visualization:
    Bar char displaying revenue  for each product is attached.
4. Insights:
    The product{} generates the highest revenue.
    On average, {:.2f} units are sold per day.

""". format(statistics,product_revenue,highest_revenue_product,average_quantity_per_day,highest_revenue_product,average_quantity_per_day)

print(report)



```

7. Saving the report to a file
---

```python
with open('reportnijake.txt','w') as f:
  f.write(report)
```
